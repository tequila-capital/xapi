```python
import os
import uuid

from dotenv import load_dotenv

from xapi_client import OrderSide, XapiClient
```

```python
load_dotenv('../.env')
server = os.getenv('SERVER')
user = os.getenv('USER')
domain = os.getenv('DOMAIN')
password = os.getenv('PASSWORD')
locale = os.getenv('LOCALE')
account = os.getenv('ACCOUNT')
route = os.getenv('ROUTE')
port = os.getenv('PORT')
```

```python
order_1_tag = f'Order-{uuid.uuid4()}'
order_1 = None

order_2_tag = f'Order-{uuid.uuid4()}'
order_2 = None
```

```python

with XapiClient(server=server,user=user,domain=domain,password=password,locale=locale,account=account,route=route,port=port, pem_path='../resources/roots.pem') as clnt:
    order_1 = clnt.submit_order(
        symbol='AAPL',
        side=str(OrderSide.BUY),
        quantity=5,
        order_tag=order_1_tag,
    )

    print(order_1)


    order_2 = clnt.submit_order(
        symbol='TSLA',
        side=str(OrderSide.BUY),
        quantity=2,
        order_tag=order_2_tag,
    )

    print(order_2)
```
